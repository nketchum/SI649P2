```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
DATA_DIR = 'csv/'
IMG_DIR = 'img/'
```

## Data

```python
monsoon = pd.read_csv(DATA_DIR + 'Monsoon_data.csv', parse_dates=['Date'])
monsoon.set_index('Date', drop=True, inplace=True)

# monsoon.head()
```

```python
olou = pd.read_csv(DATA_DIR + 'Olou_counts.csv', parse_dates=['Date'])
olou.set_index('Date', drop=True, inplace=True)

# olou.head()
```

```python
viz1 = plt.figure()

ax1 = viz1.add_subplot(211)
ax2 = viz1.add_subplot(212)

ax1.step(monsoon.index.date, monsoon.Precip, lw=1.0)
ax1.grid(True)

ax2.plot(olou.index.date, olou.Counts/1000, 'r.', ms=3.0)
ax2.grid(True)

plt.show(viz1)
```
