## BA14 Reproductions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
DATA_DIR = 'csv/'
IMG_DIR = 'img/'
```

### Data

```python
# Monsoon data.
monsoon = pd.read_csv(DATA_DIR + 'Monsoon_data.csv', parse_dates=['Date'])
monsoon.set_index('Date', drop=True, inplace=True)
```

### Exploration – Monsoons

```python
# Per month.
monsoon_ba14_monthly = monsoon
monsoon_ba14_monthly.head(200)
```

```python
# Monthly monsoons.
viz_monsoon_monthly = plt.figure()
plt.plot(monsoon_ba14_monthly.index.to_pydatetime(), monsoon_ba14_monthly['Precip'])
```

```python
# Per year.
monsoon_ba14_yearly = pd.DataFrame(monsoon_ba14_monthly.resample('Y').sum())
monsoon_ba14_yearly.head()
```

```python
# Yearly monsoons.
viz_monsoon_yearly = plt.figure()
plt.plot(monsoon_ba14_yearly.index.to_pydatetime(), monsoon_ba14_yearly['Precip'])
```

### Exploration – Olou

```python
# Olou neutron monitor counts.
olou = pd.read_csv(DATA_DIR + 'Olou_counts.csv', parse_dates=['Date'])
olou.set_index('Date', drop=True, inplace=True)
```

```python
# Per month.
olou_ba14_monthly = olou
olou_ba14_monthly.head()
```

```python
# Montly Olou counts.
viz_olou_monthly = plt.figure()
plt.plot(olou_ba14_monthly.index.to_pydatetime(), olou_ba14_monthly['Counts'])
```

```python
# Per year.
olou_ba14_yearly = pd.DataFrame(olou_ba14_monthly.resample('Y').mean())
olou_ba14_yearly.head()
```

```python
# Yearly Olou counts.
viz_olou_yearly = plt.figure()
plt.plot(olou_ba14_yearly.index.to_pydatetime(), olou_ba14_yearly['Counts'])
```
